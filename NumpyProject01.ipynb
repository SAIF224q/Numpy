Before moving to our main problem let's break it to only one dice and plot its outcome.

````python
Dice = np.random.randint(1,7,size=(1,1000))
unique_values, counts = np.unique(Dice, return_counts=True)
plt.bar(unique_values, counts)
plt.show()
````

Now let's move to our main problem

````python
Dice_1 = np.random.randint(1,7,size=(1,1000))
Dice_2 = np.random.randint(1,7,size=(1,1000))
Dice_sum = Dice_1 + Dice_2
unique_values, counts = np.unique(Dice_sum, return_counts=True)
plt.bar(unique_values, counts)
plt.show()
````
